```python
import numpy as np
import cv2
```

```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

O código acima é fundamental para disponibilizarmos uma imagem de entrada a ser processada por nosso algoritmo. A função **files.upload()** carrega o elemento de interface gráfica em que você escolhe a imagem diretamente de seu disco rígido.

Logo abaixo tem-se um laço for que percorre cada arquivo que foi subido (no caso será apenas um) imprimindo na tela a informação contendo o tamanho em bytes da imagem em questão. Esse dado é importante até para depurarmos se o upload foi feito corretamente.

```python
print(type(uploaded))
```

Temos que *uploaded* é um objeto Python do tipo dicionário (*dict*). Mais especificamente, temos que *`{'ada.jpg': 'conteúdo de cores da imagem em formato binário'}`*.

Em outras palavras, é um dicionário contendo uma única entrada. A chave em questão é o caminho + nome da imagem (*ada.jpg*) e o valor é o arquivo de imagem (cores) em representação binária.

```python
filename = next(iter(uploaded))
print(filename)
```

O método **iter(d)** retorna um iterador apontado para o dicionário *d*. Se você for entendido de linguagens antigas como C, saiba que o iterador é tipo um ponteiro, que aponta para o dicionário em questão.

Já **next(i)** retorna a próxima chave de um iterador *i*. Como esse método foi chamado uma única vez, este retornará a chave do primeiro elemento (*ada.jpg*).

```python
img = cv2.imread(filename, 0)
```

```python
from matplotlib import pyplot as plt
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

O código acima apenas exibe a imagem na tela por meio de uma janela gráfica, conforme explicado aqui.
